# Forecasting Greater Hyderabad Built-up Area Using Linear Regression (2000–2030)

This notebook analyzes the historical built-up area growth of Greater Hyderabad and applies a Linear Regression model to forecast the built-up area for the year 2030.

---
## Objectives
- Analyze historical built-up area data (2000–2024)
- Build a Linear Regression model
- Predict built-up area for 2030
- Visualize historical trends and forecast

## 1. Import Required Libraries
We import the necessary Python libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
print("Imported the necessary libraries successfully!")
```

## 2. Dataset Preparation

The dataset consists of:
- Years: 2000, 2010, 2015, 2020, 2024
- Built-up area (sq km) corresponding to each year

The data is structured for regression analysis.

```python
years = np.array([2000, 2010, 2015, 2020, 2024]).reshape(-1, 1)
area_sqkm = np.array([213.2667, 283.1418, 452.7, 581.9, 695.3])

years, area_sqkm
```

## 3. Model Development – Linear Regression

We use a Linear Regression model to understand the relationship between:
- Independent Variable (X): Year
- Dependent Variable (Y): Built-up Area

The model fits a straight line:
\[
Y = mX + c
\]

```python
model = LinearRegression()
model.fit(years, area_sqkm)

print("Slope (Growth Rate):", model.coef_[0])
print("Intercept:", model.intercept_)
```

## 4. Forecasting Built-up Area for 2030

Using the trained model, we predict the built-up area for the year 2030.

```python
# Predict 2030
future_year = np.array([[2030]])
predicted_area = model.predict(future_year)

print(f"Projected built-up area in 2030: {predicted_area[0]:.2f} sq km")
```

## 5. Visualization of Historical Trend and Forecast

The plot includes:
- Blue points → Observed data (2000–2024)
- Green dashed line → Linear trend
- Red point → 2030 forecast

```python
plt.figure(figsize=(10, 6))

# Observed data
plt.scatter(years, area_sqkm, color="blue", label="Observed (2000–2024)")

# Regression line
plt.plot(years, model.predict(years), color="green", linestyle="--", label="Linear Trend")

# 2030 prediction
plt.scatter(future_year, predicted_area, color="red", s=100, label="2030 Forecast")

plt.title("Greater Hyderabad Built-up Area Forecast (2000–2030)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Built-up Area (sq km)")
plt.grid(True, alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```

## 6. Interpretation of Results

- The built-up area shows a strong increasing trend over time.
- The Linear Regression model estimates a consistent growth rate.
- The projected built-up area for 2030 indicates continued urban expansion.

Note:
This is a simple linear model. Real-world urban growth may be influenced by:
- Policy changes
- Economic factors
- Infrastructure development
- Environmental constraints

For improved forecasting, advanced models like Polynomial Regression, ARIMA, or Machine Learning models could be explored.

## 7. Conclusion

This analysis demonstrates how Linear Regression can be used to forecast urban growth trends.

The model predicts that Greater Hyderabad will continue expanding significantly by 2030, emphasizing the need for sustainable urban planning strategies.
